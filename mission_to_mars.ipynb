```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import pprint
```

# NASA Mars News

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
url = 'https://redplanetscience.com/'
browser.visit(url)
time.sleep(5)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
mars_collection = {}

results = soup.find('section', class_='image_and_description_container')

news_title = results.find_all('div', class_='content_title')[0].text
news_description = results.find_all('div', class_='article_teaser_body')[0].text

mars_collection['news_title'] = news_title
mars_collection['news_description'] = news_description

mars_collection
```

# Scrape JPL Space Images

```python
jpl_url = 'https://spaceimages-mars.com'
browser.visit(jpl_url)
time.sleep(5)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
images = soup.find('img', class_='headerimage')
img_src = 'https://spaceimages-mars.com/' + images['src']
mars_collection['featured_image'] = img_src
mars_collection
```

# Mars Facts

```python
mars_facts_url = 'https://galaxyfacts-mars.com'

tables = pd.read_html(mars_facts_url)
df = tables[0]
html_table = df.to_html()

html_table
```

```python
mars_collection['mars_facts'] = html_table

pprint.pprint(mars_collection)
```
